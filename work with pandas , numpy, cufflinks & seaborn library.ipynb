```python
df2=sns.load_dataset('tips')
```

```python
df2
```

```python
sns.scatterplot(x='total_bill',y='tip',data=df2)
```

#relationship between 'total_bill' and 'tip'

```python
sns.relplot(x='total_bill',y='tip',data=df2)
```

# 3D-graph

```python
df2.iplot(kind='scatter3d', x = 'total_bill', y = 'tip',z ='size')
```

# A person pay hight bill who is smoker or non-smoker?

```python
sns.relplot(x = 'total_bill', y = 'tip',data=df2,hue='smoker')
```

```python
sns.relplot(x = 'total_bill', y = 'tip',data=df2,hue='smoker',style='size')
```

```python
sns.relplot(x = 'total_bill', y = 'tip',data=df2,hue='smoker',style='time')
```
